**4. Program to print Fibonacci Series upto a number**

```python

def fibonacci(n):
    a = 0
    b = 1
    if n < 0:
        print("Incorrect input")
    elif n == 0:
        print("Empty")
    elif n == 1:
        print(a)
    elif n == 2:
        print(a)
        print(b)
    else:
        print(a)
        print(b)
        for i in range(2, n):
            c = a + b
            print(c)
            a = b
            b = c

# Calling fibonacci function
print("Fibonacci Series : ")
fibonacci(10)
```
